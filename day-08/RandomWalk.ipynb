# Matplotlib Random Walk Example

Here's a simple example of a random walk I did without numpy, using vanilla Python lists for the values.  The array of x values is a simple increment.  y starts at a random value and increases or decreases by 3 each time.

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import random
import time
random.seed(time.time())
```

```python
x = [i for i in range(1,101)]

start = random.randint(0,100)
y = []
for _ in range(1,101):
    start = start + random.randint(-3,3)
    y.append(start)

plt.rcParams["figure.figsize"] = [8,4]
plt.rcParams["figure.autolayout"] = True
plt.title("Random Walk")
plt.plot(x, y, color="blue")    
plt.show()
```
